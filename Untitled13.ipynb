```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
df = pd.read_csv('/content/Iris.csv')
df.head()
```

#Step 1 — Load, Explore & Preprocess


```python
df.shape
df.info()
df.describe()
df.isnull().sum()
encoder = LabelEncoder()
df['Species'] = encoder.fit_transform(df['Species'])
X = df.drop('Species', axis=1)
y = df['Species']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

##Step 1 conclusion:

##The Iris dataset was successfully loaded and explored. No missing values were found. The categorical target variable (Species) was encoded into numerical values using LabelEncoder. The dataset was then divided into training (80%) and testing (20%) sets to prepare it for machine learning model training and evaluation.

#Step 2 — Feature Engineering

##Correlation Analysis

```python
corr_matrix = df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

##Train a Random Forest to Find Feature Importance

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
    })
importance = importance.sort_values(by='Importance',ascending=False)
print(importance)
```

##Visualize Feature Importance

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Importance',
    y='Feature',
    data=importance
)
plt.title("Feature Importance")
plt.show()
```

##Drop irrelevant features

```python
X = df.drop('SepalWidthCm', axis=1)
```

##Step 2 Conclusion:

Correlation analysis and Random Forest feature importance were used to identify the most relevant features. PetalLengthCm and PetalWidthCm were found to be the most influential predictors. SepalLengthCm and SepalWidthCm had lower importance but still contributed useful information. Therefore, all features were retained for model training to maximize predictive performance.

# Step 3 — Train 3 Different Model

##Model 1: Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
```

##Model 2: K-Nearest Neighbors (KNN)

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
```

##Model 3: Random Forest

```python
rf = RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
```

##Step 3 Conclusion

###Three classification models were trained on the Iris dataset: Logistic Regression, K-Nearest Neighbors (KNN), and Random Forest. These models were selected because they represent different machine learning approaches and are commonly used for classification tasks. The trained models will be evaluated and compared in the next step using performance metrics such as Accuracy, Precision, Recall, and F1-score.

#Step 4 — Evaluate & Compare All Models

##Evaluate Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
lr_accuracy = accuracy_score(y_test, lr_pred)
lr_precision = precision_score(y_test, lr_pred, average='weighted')
lr_recall = recall_score(y_test, lr_pred, average='weighted')
lr_f1 = f1_score(y_test, lr_pred, average='weighted')
print("Logistic Regression")
print("Accuracy:", lr_accuracy)
print("Precision:", lr_precision)
print("Recall:", lr_recall)
print("F1 Score:", lr_f1)
```

##Evaluate KNN

```python
knn_accuracy = accuracy_score(y_test, knn_pred)
knn_precision = precision_score(y_test, knn_pred, average='weighted')
knn_recall = recall_score(y_test, knn_pred, average='weighted')
knn_f1 = f1_score(y_test, knn_pred, average='weighted')
print("KNN Model")
print("Accuracy:", knn_accuracy)
print("Precision:", knn_precision)
print("Recall:", knn_recall)
print("F1 Score:", knn_f1)
```

##Evaluate Random Forest

```python
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred, average='weighted')
rf_recall = recall_score(y_test, rf_pred, average='weighted')
rf_f1 = f1_score(y_test, rf_pred, average='weighted')
print("Random Forest Model")
print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F1 Score:", rf_f1)
```

##Comparison Table

```python
comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'KNN', 'Random Forest'],
    'Accuracy': [lr_accuracy, knn_accuracy, rf_accuracy],
    'Precision': [lr_precision, knn_precision, rf_precision],
    'Recall': [lr_recall, knn_recall, rf_recall],
    'F1 Score': [lr_f1, knn_f1, rf_f1]
})
comparison
```

#Step 5 — Best Model Analysis & Conclusion

##Confusion Matrix

```python
cm = confusion_matrix(y_test, rf_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.show()
```

##Best Model Analysis
Random Forest was selected as the best model for this project. The confusion matrix shows that all test samples were classified correctly. No misclassifications were observed. The model achieved 100% Accuracy, Precision, Recall, and F1 Score. Therefore, Random Forest proved to be a highly effective classifier for the Iris dataset.
